```python
import numpy as np
import matplotlib.pyplot as plt

from multizone.src.models import insideout, twoinfall_expvar, twoinfall_linvar
```

```python
rarr = np.arange(0, 15.6, 0.1)

fig, ax = plt.subplots()
ax.plot(rarr, [insideout.timescale(r) for r in rarr], label='J21')
ax.plot(rarr, [twoinfall_expvar.timescale(r) for r in rarr], linestyle='--', label='Exponential')
ax.plot(rarr, [twoinfall_linvar.timescale(r) for r in rarr], linestyle=':', label='Linear')
ax.set_xlabel('R [kpc]')
ax.set_ylabel('tau_sfh [Gyr]')
plt.show()
```
