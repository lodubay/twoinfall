# Investigating alpha-bimodality driven by a sudden, short-lived decrease in star formation efficiency

```python
import numpy as np
import matplotlib.pyplot as plt

import vice

from multizone.src.yields import J21
vice.yields.sneia.settings['fe'] *= 10**-0.05
# from vice.yields.presets import JW20
from multizone.src.models.utils import exponential
from track_and_mdf import setup_figure, plot_vice_onezone
from apogee_sample import APOGEESample
import paths
from utils import get_bin_centers
from _globals import ONEZONE_DEFAULTS
from colormaps import paultol
```

```python
# Plot settings
plt.style.use(paths.styles / 'paper.mplstyle')
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', paultol.bright.colors)
```

Define a simple class which returns a constant SFE timescale at all times
except during a short window where the SFE timescale is much higher.

```python
class tau_star_burst:
    """
    A class declaring the time-dependence of a bursty SFE timescale.

    Attributes
    ----------
    base : float (default: 2.0)
        The initial SFE timescale in Gyr.
    onset : float (default: 4.2)
        The onset time of the sudden change in SFE timescale in Gyr.
    duration : float (default: 0.3)
        The duration of the SFE timescale change in Gyr.
    enhancement : float (default: 10.0)
        The factor to increase the SFE timescale during the hiatus.
    
    Calling
    -------
    Parameters:
        - time (float) : simulation time in Gyr.
    Returns:
        - float : the SFE timescale in Gyr.
    """
    def __init__(self, base=2, onset=4.2, 
                 duration=0.3, enhancement=10):
        self.base = base
        self.onset = onset
        self.duration = duration
        self.enhancement = enhancement

    def __call__(self, time):
        if time < self.onset:
            return self.base #* PRE_BURST_ENHANCEMENT
        elif time < self.onset + self.duration:
            return self.base * self.enhancement
        else:
            return self.base
```

```python
# Set up output directory
output_dir = paths.data / 'onezone' / 'sfe_hiatus'
if not output_dir.exists():
    output_dir.mkdir(parents=True)

# One-zone model parameters
simtime = np.arange(0, 13.21, 0.01)

# Reference: constant tau_star
const_sfe_name = str(output_dir / 'constant')
ifr = exponential(norm=1, timescale=15)
sz = vice.singlezone(
    name = const_sfe_name,
    func = ifr,
    mode = 'ifr',
    **ONEZONE_DEFAULTS
)
sz.tau_star = 2.
sz.run(simtime, overwrite=True)

# SFE hiatus model 1
sfe_hiatus1_name = str(output_dir / 'onset03_width03')
tau_star = tau_star_burst(onset=0.3)
sz = vice.singlezone(
    name = sfe_hiatus1_name,
    func = ifr,
    mode = 'ifr',
    **ONEZONE_DEFAULTS
)
sz.tau_star = tau_star
sz.run(simtime, overwrite=True)

# SFE hiatus model 2
sfe_hiatus2_name = str(output_dir / 'onset10_width03')
tau_star = tau_star_burst(onset=1)
sz = vice.singlezone(
    name = sfe_hiatus2_name,
    func = ifr,
    mode = 'ifr',
    **ONEZONE_DEFAULTS
)
sz.tau_star = tau_star
sz.run(simtime, overwrite=True)

# SFE hiatus model 3
sfe_hiatus3_name = str(output_dir / 'onset30_width03')
tau_star = tau_star_burst(onset=3)
sz = vice.singlezone(
    name = sfe_hiatus3_name,
    func = ifr,
    mode = 'ifr',
    **ONEZONE_DEFAULTS
)
sz.tau_star = tau_star
sz.run(simtime, overwrite=True)
```

A brief increase in the SFE timescale can produce a bump in the [O/Fe] distribution, 
with the timing of the hiatus controlling the location of the bump. However, 
this hiatus would have to be quite early, at ~1 Gyr, to match the APOGEE high-
alpha sequence.

```python
fig, axs = setup_figure(xlim=(-1.6, 0.6), ylim=(-0.08, 0.48))

plot_vice_onezone(const_sfe_name, fig=fig, axs=axs, label='Fiducial', marker_labels=True)
plot_vice_onezone(sfe_hiatus1_name, fig=fig, axs=axs, label='Hiatus at 0.3 Gyr')
plot_vice_onezone(sfe_hiatus2_name, fig=fig, axs=axs, label='Hiatus at 1 Gyr')
plot_vice_onezone(sfe_hiatus3_name, fig=fig, axs=axs, label='Hiatus at 3 Gyr')
axs[0].legend()

plt.show()
```

For the duration of the hiatus, gas is still falling in but star formation is
suppressed, so the Galaxy's gas mass skyrockets. When the SFE timescale drops
back to normal, this excess gas mass produces a burst in the star formation rate.

```python
fig, axs = plt.subplots(1, 4, figsize=(12, 3))
name_list = ['constant', 'onset03_width03', 'onset10_width03', 'onset30_width03']
label_list = ['Constant SFE', 'Hiatus at 0.3 Gyr', 'Hiatus at 1 Gyr', 'Hiatus at 3 Gyr']
for i, name in enumerate(name_list):
    name = str(output_dir / name)
    hist = vice.history(name)
    axs[0].plot(hist['time'], hist['ifr'], '-')
    axs[1].plot(hist['time'], hist['sfr'], '-')
    axs[2].plot(hist['time'], hist['mgas'], '-')
    axs[3].plot(hist['time'][5:], [1e-9 * mg / sfr for mg, sfr in zip(hist['mgas'], hist['sfr'])][5:], '-', label=label_list[i])
    
axs[3].legend()
for ax in axs:
    ax.set_xlabel('Time [Gyr]')
axs[0].set_ylabel(r'$\dot M_{\rm in}$')
axs[1].set_ylabel(r'$\dot M_\star$')
axs[2].set_ylabel(r'$M_{\rm gas}$')
axs[3].set_ylabel(r'$\tau_\star$')
axs[3].set_ylim((0, None))
plt.show()
```

The duration of the SFE dip controls the height of the [O/Fe] peak, because a
longer duration means more gas builds up, increasing the star formation rate
after the SFE returns to normal.

## Comparison with APOGEE data

We get a pretty good match with the data if the hiatus has a duration of
200 Myr and starts at 1.3 Gyr after the beginning of the model.

```python
feh_lim = (-1.6, 0.6)
ofe_lim = (-0.08, 0.48)
fig, axs = setup_figure(xlim=feh_lim, ylim=ofe_lim)

# Plot underlying APOGEE contours
apogee_data = APOGEESample.load()
apogee_solar = apogee_data.region(galr_lim=(7, 9), absz_lim=(0, 2))
pcm = axs[0].hexbin(apogee_solar('FE_H'), apogee_solar('O_FE'),
                gridsize=50, bins='log',
                extent=[feh_lim[0], feh_lim[1], ofe_lim[0], ofe_lim[1]],
                cmap='Greys', linewidths=0.2)
cax = axs[0].inset_axes([0.05, 0.05, 0.05, 0.8])
fig.colorbar(pcm, cax=cax, orientation='vertical', label='# APOGEE Stars')

# APOGEE abundance distributions
feh_df, bin_edges = apogee_solar.mdf(col='FE_H', range=feh_lim, 
                                        smoothing=0.2)
axs[1].plot(get_bin_centers(bin_edges), feh_df / max(feh_df), 
            color='gray', linestyle='-', marker=None)
ofe_df, bin_edges = apogee_solar.mdf(col='O_FE', range=ofe_lim, 
                                        smoothing=0.05)
axs[2].plot(ofe_df / max(ofe_df), get_bin_centers(bin_edges), 
            color='gray', linestyle='-', marker=None)

plot_vice_onezone(str(output_dir / 'constant'), fig=fig, axs=axs, 
                  label='Fiducial', marker_labels=True, linestyle='--')

# SFE hiatus model
name = str(output_dir / 'onset13_width02')
tau_star = tau_star_burst(onset=1.3, duration=0.2)
sz = vice.singlezone(
    name = name,
    func = ifr,
    mode = 'ifr',
    **ONEZONE_DEFAULTS
)
sz.tau_star = tau_star
sz.run(simtime, overwrite=True)
plot_vice_onezone(name, fig=fig, axs=axs, label='SFE Hiatus')
axs[0].legend(loc='upper right')

plt.show()
```
